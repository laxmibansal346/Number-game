# **CASE_STUDY1: Student Marks Analysis**

```
import numpy as np
marks = np.array([
[80, 75, 90],
[60, 88, 70],
[95, 92, 85],
[70, 65, 60],
[88, 78, 84]
])
#Compute total marks of each student
total = []
for i in marks:
    total.append(int(sum(i)))
print("Total marks of each student:", total)
print('\n')
#Calculate average marks per student
avg = []
for i in marks:
  avg.append(float(round(sum(i)/3,2)))
print("Average marks per student:", avg)
print('\n')
#Identify topper
topper = np.argmax(total) + 1
print("Topper student number:", topper)
print('\n')
#Find subject-wise highest score
print("Highest marks in Subject 1:", int(np.max(marks[:,0])))
print("Highest marks in Subject 2:", int(np.max(marks[:,1])))
print("Highest marks in Subject 3:", int(np.max(marks[:,2])))
print('\n')
#Compute standard deviation of each subject
print("Std of Subject 1:", round(np.std(marks[:,0]),2))
print("Std of Subject 2:", round(np.std(marks[:,1]),2))
print("Std of Subject 3:", round(np.std(marks[:,2]),2))
print('\n')

#Comment on performance consistency
print("Subjects with lower standard deviation have more consistent marks.")
```

# **CASE_STUDY2: Sales Analysis**

```
import numpy as np

sales = np.array([1200, 1500, 1100, 1700, 1600, 1800, 1400])

# 1. Average daily sales
avg = round(float(np.mean(sales)),2)
print("Average daily sales:", avg)
print("\n")
# 2. Day with highest sales
day_high = np.argmax(sales) + 1
print("Day with highest sales:", day_high)
print("\n")

# 3. Percentage growth from Day 1 to Day 7
growth = round(((sales[-1] - sales[0]) / sales[0]) * 100, 2)
print("Percentage growth from Day 1 to Day 7:", growth, "%")
print("\n")
# 4. Cumulative sales
cumulative = []
total = 0
for s in sales:
    total += s
    cumulative.append(int(total))   # convert to normal int
print("Cumulative sales:", cumulative)
print("\n")
# 5. Standard deviation
std = round(np.std(sales),2)
print("Standard deviation of sales:", std)
print("\n")
# 6. Comment
print("Sales vary daily but overall growth can be seen.")
```
