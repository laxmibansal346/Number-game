## Task 2: Indexing & Slicing (1 Mark)

Given:

A = np.arange(1, 26).reshape(5,5)

Perform:

Extract 2nd row

Extract 3rd column

Extract 2×2 submatrix (center)

Reverse last row

```python
import numpy as np

A = np.arange(1, 26).reshape(5,5)
print(A)
# Extract 2nd row
second_row = A[1]
print('\n')
print("2nd Row:",second_row)
#print(second_row)

# Extract 3rd column
third_column = A[:, 2]
print("\n3rd Column:",third_column)
#print(third_column)

# Extract 2x2 center submatrix
center_submatrix = A[1:3, 1:3]
print("\n2x2 Center Submatrix:",center_submatrix)
#print(center_submatrix)

# Reverse last row
reverse_last_row = A[4][::-1]
print("\nReversed Last Row:",reverse_last_row)
#print(reverse_last_row)
```

## Task 4: Vectorized Computation(1 Marks)

Perform element-wise multiplication
Compute dot product
Compute mean, standard deviation

```python

import numpy as np

X = np.array([1,2,3,4])
Y = np.array([5,6,7,8])

# Element-wise multiplication
element_mul = X * Y
print("Element-wise Multiplication:")
print(element_mul)

# Dot product
dot_product = np.dot(X, Y)
print("\nDot Product:")
print(dot_product)

# Mean
mean_X = np.mean(X)
print("\nMean of X:")
print(mean_X)

# Standard Deviation
std_X = np.std(X)
print("\nStandard Deviation of X:")
print(std_X)
```
