```
# Task 1: Handling Missing Values

import pandas as pd
import numpy as np

data = {
    'Age': [25, np.nan, 30, 35, np.nan],
    'Salary': [50000, 60000, np.nan, 80000, 75000]
}

df = pd.DataFrame(data)

print("Original Data:")
print(df)

print("\nMissing Values:")
print(df.isnull().sum())

df['Age'] = df['Age'].fillna(df['Age'].mean())
df['Salary'] = df['Salary'].fillna(df['Salary'].mean())

print("\nUpdated Data:")
print(df)
```

```
# Task 2: Outlier Detection

import pandas as pd
import matplotlib.pyplot as plt

# Create dataset
df = pd.DataFrame({
    'Salary': [50000, 52000, 58000, 60000, 120000]
})

# Boxplot
plt.boxplot(df['Salary'])
plt.title("Boxplot of Salary")
plt.show()

# IQR method
Q1 = df['Salary'].quantile(0.25)
Q3 = df['Salary'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

# Show outlier
print("Outlier values:")
print(df[(df['Salary'] < lower) | (df['Salary'] > upper)])

# Remove outlier
df_clean = df[(df['Salary'] >= lower) & (df['Salary'] <= upper)]

print("\nData after removing outlier:")
print(df_clean)
```

```
# Task 3: Data Scaling

import pandas as pd
from sklearn.preprocessing import StandardScaler

# Create dataset
df = pd.DataFrame({
    'Salary': [50000, 52000, 58000, 60000]
})

# Apply scaling
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)

# Convert to DataFrame for better display
scaled_df = pd.DataFrame(scaled_data, columns=['Salary'])

# Display scaled data
print("Scaled Data:")
print(scaled_df)
```
