```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.DataFrame({
    'Age': [22, 25, 47, 52, 46],
    'Income': [15000, 18000, 65000, 72000, 69000]
})

# Histogram
plt.figure(figsize=(3,2))
plt.hist(df['Age'])
plt.show()
print("\n")
# Interpretation: Most values are in middle range.


# Boxplot
plt.figure(figsize=(3,2))
plt.boxplot(df['Income'])
plt.show()
print("\n")
# Interpretation: No big outlier present.


# Heatmap
plt.figure(figsize=(3,2))
sns.heatmap(df.corr(), annot=True)
plt.show()
print("\n")
# Interpretation: Age and Income are related.


# Scatter Plot
plt.figure(figsize=(3,2))
plt.scatter(df['Age'], df['Income'])
plt.show()
print("\n")
# Interpretation: Income increases with age.


# Bar Chart
plt.figure(figsize=(3,2))
plt.bar(df['Age'], df['Income'])
plt.show()
# Interpretation: Higher age → higher income.
```

```python
import pandas as pd
from scipy import stats

# t-test
data1 = [10, 12, 14, 16, 18]
data2 = [8, 9, 11, 13, 15]

t, p = stats.ttest_ind(data1, data2)
print("p-value (t-test):", p)

if p < 0.05:
    print("Reject H0")
else:
    print("Accept H0")


# chi-square test
data = [[20, 30], [10, 40]]

chi, p, d, exp = stats.chi2_contingency(data)
print("\np-value (chi-square):", p)

if p < 0.05:
    print("Reject H0")
else:
    print("Accept H0")
```

#Interpretation:
# For t-test, p-value is greater than 0.05 so there is no significant difference.

# For chi-square test, p-value is less than 0.05 so variables are related.
